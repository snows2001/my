#클래스를 구경해 봐요!

## 자바에서 클래스란?

>**클래스**는 **어떤 종류**의 **데이터**와 **데이터를 다루는 방법**들을  모아둔 것입니다.

이 말이 무슨 말이냐?  우리 하나씩 파헤쳐 가보죠.

위 문장에서 **어떤 종류**라 함은 **여러분 머리 속에 떠오르는 모든 것**입니다. **보이는 것들**로 예를 들면 핸드폰, 자동차, 음식과 같은 모든 것이 해당됩니다. **안 보이는 것들**로 예를 들면 사랑, 꿈, 생각과 같은 모든 것이 해당됩니다.
<br />
<br />

**<q>클래스로 만들 수 있는 것은 세상의 모든 것입니다.</q>**

지금은 없지만 여러분들이 상상해 내는 모든 것들도 포함됩니다.

자바는 세상의 모든 것을 프로그램 속으로 넣을 수 있는 마법의 언어입니다.

이제 클래스를 배웠으니 만들어보죠.













** 보이는 것들: ** 

핸드폰 클래스: 핸드폰을 관리하기 위한 클래스입니다. 핸드폰의 제조사, 모델명, 용량, 화면 크기, 배터리 수명 등을 속성으로 가지며, 전화를 거는, 문자를 보내는, 인터넷을 사용하는 등의 메서드를 포함합니다.

자동차 클래스는 자동차의 정보와 기능을 관리하기 위한 클래스입니다. 예를 들어, 자동차의 모델명, 연식, 최고 속도, 최대 탑승인원 등과 같은 정보를 속성으로 가지고, 주행을 시작하거나 멈추는 메서드를 포함합니다

음식 클래스: 음식을 관리하기 위한 클래스입니다. 음식의 종류, 이름, 가격, 칼로리, 원산지 등을 속성으로 가지며, 주문을 받는, 조리하는, 서빙하는 등의 메서드를 포함합니다.


** 안 보이는 것들: ** 

사랑 클래스: 이 클래스는 사랑과 관련된 데이터와 함수를 모아놓은 것입니다. 예를 들어, 이 클래스는 "사랑하는 사람의 이름", "사랑하는 사람과의 관계", "사랑하는 방식" 등과 같은 데이터를 포함할 수 있습니다. 이 클래스에는 또한 "사랑하는 사람에게 선물을 줄 때의 행동", "사랑하는 사람을 위해 할 일", "사랑하는 사람과의 데이트 계획" 등과 같은 함수를 포함할 수 있습니다.

꿈 클래스: 이 클래스는 꿈과 관련된 데이터와 함수를 모아놓은 것입니다. 예를 들어, 이 클래스는 "꿈을 꾼 시간", "꿈에서 본 내용", "꿈을 꾸는 동안의 감정" 등과 같은 데이터를 포함할 수 있습니다. 이 클래스에는 또한 "꿈 해석", "꿈을 기록하는 기능", "꿈에 나온 인물과의 관계 탐색" 등과 같은 함수를 포함할 수 있습니다.

생각 클래스: 이 클래스는 개인이 가지고 있는 생각과 관련된 데이터와 함수를 모아놓은 것입니다. 예를 들어, 이 클래스는 "생각한 시간", "생각한 내용", "생각할 때의 감정" 등과 같은 데이터를 포함할 수 있습니다. 이 클래스에는 또한 "생각 정리 기능", "생각에 대한 기록 기능", "생각에 대한 탐구 기능" 등과 같은 함수를 포함할 수 있습니다.

```java
public class Phone { // Phone 클래스 선언

    // Phone 클래스의 속성(멤버 변수) 선언
    String model; // 모델명
    int price; // 가격
    boolean isOn; // 전원 상태

    // Phone 클래스의 메서드(멤버 함수) 선언
    public void turnOn() { // 전원을 켜는 메서드
        isOn = true; // 전원 상태를 켜진 상태로 변경
        System.out.println("전원을 켭니다."); // "전원을 켭니다." 출력
    }

    public void turnOff() { // 전원을 끄는 메서드
        isOn = false; // 전원 상태를 꺼진 상태로 변경
        System.out.println("전원을 끕니다."); // "전원을 끕니다." 출력
    }

    public void call(String number) { // 전화를 거는 메서드
        System.out.println(number + "번으로 전화를 겁니다."); // 입력한 번호로 전화를 걸고 "번호 + 번으로 전화를 겁니다." 출력
    }

    public void sendMessage(String number, String message) { // 메시지를 보내는 메서드
        System.out.println(number + "번으로 메시지를 보냅니다: " + message); // 입력한 번호로 메시지를 보내고 "번호 + 번으로 메시지를 보냅니다: 메시지 내용" 출력
    }
}
```

자바로 게임을 만들 때는 플레이어, 적, 아이템, 무기 등의 클래스가 필요합니다. 

플레이어 클래스의 **데이터**는 플레이어의 이름, 체력, 공격력 등이 해당됩니다. **데이터를 처리하는 방법**은 공격하거나 회복하는 등의 행동과 관련된 메서드가 해당됩니다.

앞 문장에서 "데이터"는 클래스가 가지는 속성(attribute)으로, 플레이어 클래스에서는 플레이어의 이름, 체력, 공격력 등이 해당됩니다. "데이터를 처리하는 방법"은 클래스가 가지는 메서드(method)로, 플레이어 클래스에서는 공격하거나 회복하는 등의 행동과 관련된 메서드가 해당됩니다.

예를 들어, 우리가 만들려는 게임 프로그램에서 캐릭터를 표현하기 위해 Character 클래스를 만들 수 있습니다. Character 클래스는 캐릭터의 이름, 레벨, 경험치 등과 같은 데이터를 포함하며, 캐릭터를 공격하거나 아이템을 사용하는 등의 동작을 처리하는 메서드를 포함할 수 있습니다. 이렇게 Character 클래스는 캐릭터의 데이터와 행동을 모아둔 것입니다.

다른 예로는 예를 들어 "사람" 클래스가 있을 수 있습니다. 이 클래스는 "이름", "나이", "성별" 등과 같은 데이터를 포함하고, "인사하기", "먹기", "걷기"와 같은 동작을 처리하는 방법을 모아둘 수 있습니다. 또 다른 예로는 "자동차" 클래스가 있을 수 있는데, 이 클래스는 "차종", "색상", "연식"과 같은 데이터와 "시동 걸기", "가속하기", "제동하기"와 같은 동작을 처리하는 방법을 모아둘 수 있습니다.

게임 프로그램: 게임을 만들 때는 플레이어, 적, 아이템, 무기 등의 클래스가 필요합니다. 각 클래스는 그에 해당하는 데이터와 메서드를 가지고 있어서 게임 내에서 사용됩니다.

은행 프로그램: 은행 프로그램을 만들 때는 계좌, 입출금, 이체, 대출 등의 클래스가 필요합니다. 계좌 클래스는 계좌번호, 잔액 등의 데이터와 입금, 출금 메서드를 가지고 있고, 입출금 클래스는 입금, 출금 메서드를 가지고 있습니다.
